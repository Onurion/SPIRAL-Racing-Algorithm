```python
import numpy as np

def calculate_stats(all_data):
    # Calculate mean and standard deviation for each drone
    drone_stats = {}
    for typ, data in all_data.items():
        drone_stats[typ] = {}
        for drone, episode_data in data.items():
            lap_times = episode_data['lap_time']
            mean_lap_time = np.mean(lap_times)
            std_lap_time = np.std(lap_times)

            if 'successful_flight' in list(episode_data.keys()): 

                successful_flights = episode_data['successful_flight']
                num_successful_flights = sum(successful_flights)
                total_flights = len(successful_flights)
                success_rate = num_successful_flights / total_flights

                drone_stats[typ][drone] = {
                    'mean_lap_time': mean_lap_time,
                    'std_lap_time': std_lap_time,
                    'num_successful_flights': num_successful_flights,
                    'total_flights': total_flights,
                    'success_rate': success_rate
                }

            else:
                drone_stats[typ][drone] = {
                    'mean_lap_time': mean_lap_time,
                    'std_lap_time': std_lap_time,
                }


        # Print the results for each drone
        for drone, stats in drone_stats[typ].items():
            print(f"{drone}:")
            print(f"  Mean Lap Time: {stats['mean_lap_time']:.3f}")
            print(f"  Standard Deviation of Lap Time: {stats['std_lap_time']:.3f}")

            if 'successful_flight' in list(episode_data.keys()): 
                print(f"  Number of Successful Flights: {stats['num_successful_flights']}")
                print(f"  Total Flights: {stats['total_flights']}")
                print(f"  Success Rate: {stats['success_rate']:.3f}")


        # Calculate mean and standard deviation of the means and standard deviations
        mean_lap_times = [stats['mean_lap_time'] for stats in drone_stats[typ].values()]
        std_lap_times = [stats['std_lap_time'] for stats in drone_stats[typ].values()]

        if 'successful_flight' in list(episode_data.keys()): 
            success_rates = [stats['success_rate'] for stats in drone_stats[typ].values()]
            mean_of_means_success = np.mean(success_rates)
            std_of_means_success = np.std(success_rates)

            drone_stats[typ]['ultimate_mean_success'] = mean_of_means_success
            drone_stats[typ]['ultimate_std_success'] = std_of_means_success

        

        mean_of_means = np.mean(mean_lap_times)
        std_of_means = np.std(mean_lap_times)
        mean_of_stds = np.mean(std_lap_times)
        std_of_stds = np.std(std_lap_times)

        drone_stats[typ]['ultimate_mean'] = mean_of_means
        drone_stats[typ]['ultimate_std'] = std_of_means
    
        

        # Print the overall results
        print(f"\nOverall {typ}: ")
        print(f"  Mean of Mean Lap Times: {mean_of_means:.3f}")
        print(f"  Standard Deviation of Mean Lap Times: {std_of_means:.3f}")
        print(f"  Mean of Standard Deviations: {mean_of_stds:.3f}")
        print(f"  Standard Deviation of Standard Deviations: {std_of_stds:.3f}")

        if 'successful_flight' in list(episode_data.keys()): 
            print(f"  Mean of Success Rates: {mean_of_means_success:.3f}")
            print(f"  Standard Deviation of Success Rates: {std_of_means_success:.3f}")
        print ("\n")

    return drone_stats


# drone_stats = calculate_stats(data_normal)
```
